```python
import pandas as pd

df = pd.read_csv(
    "/content/Project Description Agriculture PUE SMi(Agriculture GP project with PUE).csv",
    encoding='latin1'   # or try 'utf-8-sig' if this doesn't work
)

df.head()
```

```python
df = df.drop(columns=["PUE Description"])
```

```python
df.head()
```

```python
import pandas as pd
import re

# Clean column names
df.columns = df.columns.str.strip()

# Clean and convert Total PUE investment column
def sum_multiple_amounts(x):
    nums = re.findall(r"[\d\.]+", str(x))
    nums = [float(n) for n in nums]
    return sum(nums)

df["Total PUE investment (cleaned)"] = df["Total PUE investment"].apply(sum_multiple_amounts)

# Drop PUE Description if still there
if "PUE Description" in df.columns:
    df = df.drop(columns=["PUE Description"])
```

```python
import pandas as pd
import re

# Clean column names
df.columns = df.columns.str.strip()

# Clean and convert Total PUE investment column
def sum_multiple_amounts(x):
    nums = re.findall(r"[\d\.]+", str(x))
    nums = [float(n) for n in nums]
    return sum(nums)

df["Total PUE investment (cleaned)"] = df["Total PUE investment"].apply(sum_multiple_amounts)

# Drop PUE Description if still there
if "PUE Description" in df.columns:
    df = df.drop(columns=["PUE Description"])
```

```python
country_pue = (
    df.groupby("COUNTRY")["Total PUE investment (cleaned)"]
    .sum()
    .reset_index()
    .sort_values(by="Total PUE investment (cleaned)", ascending=False)
)

country_pue.head()
```

```python
import plotly.express as px

fig = px.choropleth(
    country_pue,
    locations="COUNTRY",                     # country names
    locationmode="country names",            # use country names (not ISO codes)
    color="Total PUE investment (cleaned)",  # color scale by value
    hover_name="COUNTRY",                    # tooltip label
    color_continuous_scale="Viridis",        # color theme
    title="🌍 Cumulative PUE Investment by Country (in Million USD)"
)

fig.update_layout(
    geo=dict(showframe=False, showcoastlines=True),
    coloraxis_colorbar=dict(title="Investment (Million USD)")
)

fig.show()
```

```python
# Save as CSV
country_pue.to_csv("cumulative_PUE_by_country.csv", index=False)

# Download to your computer
from google.colab import files
files.download("cumulative_PUE_by_country.csv")
```
